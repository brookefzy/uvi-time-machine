```python
# 1.2
# load all city and source of years
import os
import pandas as pd
import numpy as np

ROOT = r"D:\Dropbox (Personal)\Personal Work\_Projects2023\01_city-never-was"

# ROOT = "/Users/yuan/Dropbox (Personal)/Personal Work/_Projects2023"
EXPORT_FOLDER = f"{ROOT}/_data/_curated/c_analysis"

city_path = os.path.join(EXPORT_FOLDER,
                            f"c_city_built_environment_full_cluster=all_ncat=27.csv")
citydf = pd.read_csv(city_path)
```

```python
update_folder = f"{ROOT}/_data/_raw/_road_fatality/_fatality/"
```

```python
fatality_df = pd.read_excel(f"{ROOT}/_data/_raw/_road_fatality/_fatality/data_source_update.xlsx")
fatality_df['city_lower'] = fatality_df['City'].fillna("").str.lower().apply(lambda x: x.replace(" ", ""))
fatality_df.head()
```

```python
updatesource = citydf[['city_lower','country','num_person_killed_per_lakh_city','num_hex']]\
    .merge(fatality_df, 
           on = 'city_lower',
           how = 'left').drop_duplicates()
updatesource['Year Covered'] = np.where(
    updatesource['country']=='Brazil', 2019, updatesource['Year Covered'])
updatesource['Year Covered'] = np.where(
    updatesource['country']=='India', 2019, updatesource['Year Covered'])
updatesource['Year Covered'] = np.where(
    updatesource['country']=='India', 2019, updatesource['Year Covered'])
updatesource['Region'] = updatesource['Region'].fillna(updatesource['country'])
updatesource['City'] = updatesource['City'].fillna(updatesource['city_lower'].str.capitalize())
updatesource['Year Covered'] = np.where(
    updatesource['country']=='United States of America', 2019, updatesource['Year Covered'])
updatesource['Region'] = updatesource['Region'].fillna("United States of America")
updatesource = updatesource[~updatesource['city_lower'].isin(['lima', 'chicago'])].reset_index(drop = True)

updatesource['Source'] = np.where(
    updatesource['country']=='Brazil',
    "\cite{malta2022road}",
    updatesource['Source']
)
ussource = "\cite{usa_all_cities}"
indiasource = "\cite{india_all_cities}"
updatesource['Source'] = np.where(
    updatesource['country']=='United States of America',
    ussource,
    updatesource['Source']
)
updatesource['Source'] = np.where(
    updatesource['country']=='India',
    indiasource,
    updatesource['Source']
)
updatesource['Region'] = np.where(
    updatesource['city_lower']=='toledo',
    "United States of America",
    updatesource['Region']
)

updatesource['Year Covered'] = updatesource['Year Covered'].fillna(2019)
updatesource['Source'] = updatesource['Source'].fillna(ussource)

updatesource[['Region', 
              'City', 
              'Year Covered', 
              'num_person_killed_per_lakh_city',
              'Source']].sort_values(['Region','City']).to_csv(os.path.join(update_folder, "city_fatality_source_update.csv"), index = False)
```
